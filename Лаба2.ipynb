```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from bokeh.plotting import output_file, figure, show
from bokeh.io import output_notebook

output_notebook()

url = "https://drive.google.com/uc?export=download&id=18VeFLUIj22hRmc_tNGjfoawyn7iyqQ_w"
input_df = pd.read_csv(url, sep = ',', index_col = 0)
print(input_df.head())
print(input_df.tail())

input_matrix = input_df.values
axes = input_df.plot(marker='.', figsize=(20, 20), subplots = True)
generation_sum = (input_matrix[:144, 1:]).sum(axis = 0)
procent = (generation_sum/generation_sum.sum()) * 100
labels = input_df.columns[1:]
fig1, ax1 = plt.subplots(figsize = (11, 11))
a1 = ax1.pie(generation_sum, labels=labels, autopct='%1.1f%%')
ax1.axis('equal')
plt.savefig('диаграмма')
data_frame = pd.DataFrame({
    '1': labels,
    '2': procent
    })
data_frame.to_csv('отношение.csv', index = False, header = ['Тип станции', 'Доля выработки энергии, %'])

plt.show()
```
